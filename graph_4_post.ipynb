```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
na = pd.read_csv('NA.csv')
eu = pd.read_csv('EU.csv')
stats = [na['Flamed'].sum(), eu['Flamed'].sum()] 
y_pos = np.arange(len(stats))
```

## Wins are recorded as 1, losses - 0 
## If I got flamed in that game - its 1, if moderately flamed or not flamed at all - 0

## NA recorded results 


```
na
```

## EU recorded results


```
eu
```

## Graph


```
plt.xkcd()
plt.figure(figsize = (5, 7))
plt.bar(y_pos, stats, edgecolor='black')
plt.ylabel('Times flamed')
plt.xlabel('Server')
plt.xticks(y_pos, ['NA', 'EU'])
plt.ylim(0, 100)

for i in range(len(y_pos)):
  plt.text(x = y_pos[i]-0.07 , y = stats[i]-7.1, s = stats[i], size = 16)

plt.title('Quantity of games where I got flamed \n for playing widow (per 100 games)')
plt.show()
```

## Correlations


```
# absolute values are taken coz wins are marked as 1 
# we want to know correlation not between wins/flamed but rather losses/flamed

corr_EU = round(abs(np.corrcoef(eu['Win/Loss'], eu['Flamed'])[0, 1]), 2)
corr_NA = round(abs(np.corrcoef(na['Win/Loss'], na['Flamed'])[0, 1]), 2)

print(f'Correlation between losing a game and being flamed on EU server - {corr_EU}')
print(f'Correlation between losing a game and being flamed on NA server - {corr_NA}')
```
